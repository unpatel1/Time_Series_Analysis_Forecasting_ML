### 11 - Power, Log and Box-Cox Transformations
Date: 9/13/2022

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import boxcox
```

```python
df = pd.read_csv('international-airline-passengers.csv', index_col='Month', parse_dates=True)
```

```python
df.columns = ['Passengers']
```

```python
df.tail()
```

```python
# Removing the last row
df = df[:-1]
df.tail()
```

```python
df['Passengers'].plot(figsize = (15,6))
```

**Square root transform**

```python
# Trying square root transform
df['SqrtPassengers'] = np.sqrt(df['Passengers'])
```

```python
df['SqrtPassengers'].plot(figsize = (15,6))
```

The time series has squeezed down slightly after square root transform. But the amplitude of the seasonal pattern still increases over time.

**Log transform**

```python
# Trying log transform
df['LogPassengers'] = np.log(df['Passengers'])
```

```python
df['LogPassengers'].plot(figsize = (15,6))
```

The time series has squeezed down significantly after log transform.

**BoxCox transform**  
* Takes 1D dataset as input  
* Returns transformed data
* Returns with the optimal value of lambda

```python
data, lam = boxcox(df['Passengers'])
```

```python
# Checking the value of labmda
lam
```

```python
df['BoxCoxPassengers'] = data
df['BoxCoxPassengers'].plot(figsize = (15, 6))
```

This looks something like in between square root and log transforms.

### 12 - Forecasting metrics

Sum of Squared Errors (SSE)  
* As the # of samples increases, SSE increases too much.  
* Difficult to compare 2 datasets with both have different # of samples.  

Mean Squared Error (MSE)  
* \# of samples doesn't affect too much.  
* It is an estimate of the expected squared error. For many algo, their objective is to minimize this expected value.  

Root Mean Squared Error (RMSE)  
* SSE & MSE don't have intuitive units. e.g. data is about temp in degree kelvin, units of SSE & MSE would be kelvin squared.  
* To correct this problem, square root of MSE is taken in RMSE. So RSME for temperature data would be kelvin and not kelvin squared.  

Mean Absolute Error (MAE)  
* Corresponds to optimizing a Laplace-distributed error
* Sometimes, people train their model by using the MSE, but report the MAE as a matric.
* Some libraries / methods (e.g. ARIMA) won't give a choice what to minimize.

**Scale Invariance**  
* MSE & MAE depends on the scale of the data  
* e.g. dailt stock return in a fraction of a percent V/S house prices in 100,000s to 1,000,000s.  

$R^2$: A common matric that is sacle invariant  
* It is not an error  
* We want it to be bigger, not smaller  
* E = 1 - SSE / SST (total sum of squares)
* If we divide SSE & SST by N, we will get the following  
    * E = 1 - MSE / Sample variance (V)
* If the model makes a perfect prediction, MSE = 0 and $R^2$ = 1
* If the model is bad and can predict only the avg of the target, MSE = Sample variance (V) and $R^2$ = 0  
* $R^2$ < 0: Happens when your predictions are worse than just predicting y bar (mean of the target). 
* In that case SSE > SST  

**Scikit-Learn**  
* For classification models: model.score(X,Y) returns accuracy  
* For regression models: model.score(X,Y) returns $R^2$  

Mean Absolute Percentage Error (MAPE)  
* In order to get a scale-invariant error, express it as a %  
* It is not symmetric e.g. y = 10 & $\widehat{y}$ = 11 VS y = 11 & $\widehat{y}$ = 10  

Symmetric Mean Absolute Percentage Error (sMAPE)  
* Uses the avg of y & $\widehat{y}$ in the denominator  

**Problem with MAPE & sMAPE**  
* When the denominator is 0, error explodes to infinity.  
* Makes no sense - error should only explode if the prediction is actually wrong

### 14 - Price Simulations in Code

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# number of time steps
T = 1000

# initial price
P0 = 10

# drift
mu = 0.001
```

```python
# last log price
last_p = np.log(P0)

log_returns = np.zeros(T)
prices = np.zeros(T)

for t in range(T):
    # sample a log return
    r = 0.01 * np.random.randn()
    
    # compute the new log price
    p = last_p + mu + r
    
    # store the return and price
    log_returns[t] = r + mu
    prices[t] = np.exp(p)
    
    # assign last_p
    last_p = p
```

```python
plt.figure(figsize = (12, 5))
plt.plot(prices)
plt.title('Simulated Prices')
```
